```python
df_algo = data_plot[data_plot["Algorithm"] == "Natural Merge Sort"]

# Create a figure with four subplots
# fig, axs = plt.subplots(3,1, figsize=(6,10))
fig, axs = plt.subplots(1,3, figsize=(14,4))

# Loop through the subplots and plot the data
for z in df_algo["Data Order"].unique():
    plotter = df_algo[df_algo["Data Order"]==z]
    axs[0].plot(plotter["Data Size"], plotter["Comparisons"], "o",label=z)
    axs[1].plot(plotter["Data Size"], plotter["Exchanges"], "o",label=z)
    axs[2].plot(plotter["Data Size"], plotter["Time (sec)"], "o",label=z)


# Adjust the layout and spacing of the subplots
axs[0].set_title("Comparisons")
axs[1].set_title("Exchanges")
axs[2].set_title("Time (sec)")
fig.suptitle("Algorithm: Natural Merge Sort")
plt.tight_layout()
fig.subplots_adjust(top=0.84)
plt.legend()
```
